```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from custom_io import *
```

```python
trial5 = read_stats("logs/1x9-trial5/1620405842.csv", True, ",")
print(trial5)
```

```python
trial5[:, 0]
```

```python
trial5_min = read_stats("logs/1x9-trial5/1620405842-2.csv", True, ",")
print(trial5_min)
```

```python
trial5_max = read_stats("logs/1x9-trial5/1620405842-3.csv", True, ",")
print(trial5_max)
```

```python
plt.figure(figsize = (8, 8))
#plt.plot(trial5[:, 1], trial5[:, 2], color = 'blue', label = "Other")
plt.plot(trial5[:, 1], [avg(trial5[:a, 2]) for a in range(1, len(trial5[:, 2]) + 1)], label = "Average Reward")
plt.plot(trial5_min[:, 1], [avg(trial5_min[:a, 2]) for a in range(1, len(trial5_min[:, 2]) + 1)], label = "Min Reward")
plt.plot(trial5_max[:, 1], [avg(trial5_max[:a, 2]) for a in range(1, len(trial5_max[:, 2]) + 1)], label = "Max Reward")
#plt.plot(fit_x, fit_y, color = 'green', label = 'Cubic Fit')
plt.ylabel(r'Reward')
plt.xlabel('Episode Number')
plt.title("Incremental Average of Trial5 Rewards")
plt.legend()
plt.style.use('ggplot')
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=0.8
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'
plt.savefig('trial5_inc.png')
plt.show()
```

```python
plt.figure(figsize = (8, 8))
#plt.plot(trial5[:, 1], trial5[:, 2], color = 'blue', label = "Other")
plt.plot(trial5[:, 1], trial5[:, 2], label = "Average Reward")
plt.plot(trial5_min[:, 1], trial5_min[:, 2], label = "Min Reward")
plt.plot(trial5_max[:, 1], trial5_max[:, 2], label = "Max Reward")
#plt.plot(fit_x, fit_y, color = 'green', label = 'Cubic Fit')
plt.ylabel(r'Reward')
plt.xlabel('Episode Number')
plt.title("Trial5 Rewards")
plt.legend()
plt.style.use('ggplot')
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=0.8
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'
plt.savefig('trial5_avg.png')
plt.show()
```
